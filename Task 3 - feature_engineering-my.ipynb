```python
# Create ratio features
df['ratio_elec_gas'] = df['cons_12m'] / df['cons_gas_12m'].replace(0, 1)
df['ratio_forecast_actual'] = df['forecast_cons_12m'] / df['cons_12m'].replace(0, 1)

# Create difference features
df['diff_forecast_actual'] = df['forecast_cons_12m'] - df['cons_12m']
df['diff_price_peak_offpeak'] = df['forecast_price_energy_peak'] - df['forecast_price_energy_off_peak']

# Create interaction terms
df['interaction_peak_offpeak'] = df['forecast_price_energy_peak'] * df['forecast_price_energy_off_peak']
df['interaction_pow_peak_offpeak'] = df['forecast_price_pow_off_peak'] * df['forecast_price_energy_peak']

# Show the first few rows of the new DataFrame
print(df.head())
```

```python
price_df = pd.read_csv('price_data (1).csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
df1 = pd.read_csv('clean_data_after_eda.csv')
```

```python
# Merge cleaned_df and price_df on 'id'
merged_df = pd.merge(df1, price_df, on='id')

# Create combined features
# Overall price metrics
merged_df['overall_price_var'] = (merged_df['price_off_peak_var'] + merged_df['price_peak_var'] + merged_df['price_mid_peak_var']) / 3
merged_df['overall_price_fix'] = (merged_df['price_off_peak_fix'] + merged_df['price_peak_fix'] + merged_df['price_mid_peak_fix']) / 3

# Consumption and price interaction
merged_df['cons_price_interaction'] = merged_df['cons_12m'] * merged_df['overall_price_var']

# Aggregated forecast price
merged_df['avg_forecast_price'] = (merged_df['forecast_price_energy_off_peak'] + merged_df['forecast_price_energy_peak'] + merged_df['forecast_price_pow_off_peak']) / 3

# Show the first few rows of the new DataFrame
print(merged_df.head())
```

## 3. Feature engineering

### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```
